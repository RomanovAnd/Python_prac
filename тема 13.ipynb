```python
import matplotlib.pyplot as plt
import scipy.stats as sps
import seaborn as sb
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
n = 50
k = 5
```

## $\mathcal{N}(a_i,1)$

```python
#1
pv1 = []
pv2 = []
for i in range(100):
    a1 = sps.norm().rvs(size = n)
    a2 = sps.norm().rvs(size = n)
    a3 = sps.norm().rvs(size = n)
    a4 = sps.norm().rvs(size = n)
    a5 = sps.norm().rvs(size = n)
    pv1.append(sps.f_oneway(a1,a2,a3,a4,a5)[1])
    pv2.append(sps.kruskal(a1,a2,a3,a4,a5)[1])
plt.figure(figsize = (6,6))
plt.plot(sorted(pv1),list(range(100)),label = 'ftest')
plt.plot(sorted(pv2),list(range(100)),label = 'kruskal')
plt.legend()
plt.show()
```
